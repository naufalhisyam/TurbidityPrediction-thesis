```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

data = {'MAE': [5.13,2.44,5.12,2.59],
        'MSE': [43.31,10.00,45.45,11.45],
        'R2': [0.967,0.992,0.966,0.991]}
df= pd.DataFrame(data,columns=['MAE','MSE','R2'],
                 index = ['ResNet-50_noTL','ResNet-50_withTL','DenseNet-121_noTL','DenseNet-121_withTL'])

x_locs = np.arange(len(df['MAE']))
index=['ResNet-50_noTL','ResNet-50_withTL','DenseNet-121_noTL','DenseNet-121_withTL']

fig, ax = plt.subplots(figsize =(6, 4))
fig.suptitle(r'Perbandingan Performa Model untuk Citra Formazin dengan Sudut 0$\degree$', fontsize=10, fontweight='bold', y=0.95)

bar_size = 0.25
padding = 0.25

y_locs = np.arange(len(df['MAE'])) * (bar_size * 3 + padding)

rects1 = ax.barh(y_locs, df['MAE'], align='edge', height=bar_size, label='MAE')
rects2 = ax.barh(y_locs + bar_size, df['MSE'], align='edge', height=bar_size, label='MSE')
rects3 = ax.barh(y_locs + 2*bar_size, df['R2'], align='edge', height=bar_size, label='$R^2$')

ax.bar_label(rects1, )
ax.bar_label(rects2, )
ax.bar_label(rects3, )

ax.set(yticks=x_locs+1.5*padding, yticklabels=index,ylim=[0 - padding, len(x_locs)] )
ax.set_axisbelow(True)
ax.set_xlim([0,55])
ax.grid(color='gray', linestyle='dashed',axis='x',which='both')
ax.legend()

plt.tight_layout()
plt.savefig(f'comparPlot_90.png', dpi=150)
plt.show()
```
